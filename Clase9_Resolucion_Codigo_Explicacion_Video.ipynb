```python
from datetime import datetime

class Tarea:
    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    def mostrar(self):
        return f"Nombre: {self.nombre}\nDescripción: {self.descripcion}\nPrioridad: {self.prioridad}"

class TareaGeneral(Tarea):
    def __init__(self, nombre, descripcion, prioridad):
        super().__init__(nombre, descripcion, prioridad)

class TareaConFecha(Tarea):
    def __init__(self, nombre, descripcion, prioridad, fecha):
        super().__init__(nombre, descripcion, prioridad)
        self.fecha = fecha

    def mostrar(self):
        return super().mostrar() + f"\nFecha: {self.fecha.strftime('%d/%m/%Y')}"

class ListaDeTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)
        print(f"Tarea '{tarea.nombre}' agregada.")

    def eliminar_tarea(self, indice):
        if 0 <= indice < len(self.tareas):
            tarea = self.tareas.pop(indice)
            print(f"Tarea '{tarea.nombre}' eliminada.")
        else:
            print("Índice de tarea inválido.")

    def mostrar_tareas(self):
        if self.tareas:
            print("Lista de tareas:")
            for i, tarea in enumerate(self.tareas):
                print(f"{i+1}. {tarea.mostrar()}\n")
        else:
            print("No hay tareas pendientes.")

def main():
    lista = ListaDeTareas()
    
    while True:
        print("\n1. Agregar tarea general")
        print("2. Agregar tarea con fecha")
        print("3. Eliminar tarea")
        print("4. Mostrar tareas")
        print("5. Salir")
        
        opcion = input("Seleccione una opción: ")
        
        if opcion in ["1", "2"]:
            nombre = input("Ingrese el nombre de la tarea: ")
            descripcion = input("Ingrese la descripción de la tarea: ")
            while True:
                prioridad = input("Ingrese la prioridad (alta/media/baja): ").lower()
                if prioridad in ['alta', 'media', 'baja']:
                    break
                print("Prioridad inválida. Por favor, ingrese 'alta', 'media' o 'baja'.")
            
            if opcion == "1":
                tarea = TareaGeneral(nombre, descripcion, prioridad)
            else:
                while True:
                    fecha_str = input("Ingrese la fecha (DD/MM/AAAA): ")
                    try:
                        fecha = datetime.strptime(fecha_str, "%d/%m/%Y")
                        break
                    except ValueError:
                        print("Formato de fecha inválido. Use DD/MM/AAAA.")
                tarea = TareaConFecha(nombre, descripcion, prioridad, fecha)
            
            lista.agregar_tarea(tarea)
        elif opcion == "3":
            indice = int(input("Ingrese el número de la tarea a eliminar: ")) - 1
            lista.eliminar_tarea(indice)
        elif opcion == "4":
            lista.mostrar_tareas()
        elif opcion == "5":
            print("¡Hasta pronto!")
            break
        else:
            print("Opción inválida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```
